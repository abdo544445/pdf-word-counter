```python
import PyPDF2
from transformers import pipeline

def extract_text_from_pdf(pdf_path):
    # Initialize the AI-based OCR pipeline
    ocr = pipeline("image-to-text", model="microsoft/trocr-base-handwritten")

    # Open the PDF file
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        
        extracted_text = ""
        
        for page_num in range(num_pages):
            page = reader.pages[page_num]
            
            # Extract text using PyPDF2
            text = page.extract_text()
            
            # If PyPDF2 fails to extract text, use AI-based OCR
            if not text.strip():
                # Note: In a real implementation, you'd need to convert the PDF page to an image here
                # For demonstration, we'll assume this step and use a placeholder
                image = f"page_{page_num}.png"  # Placeholder
                text = ocr(image)[0]['generated_text']
            
            extracted_text += f"Page {page_num + 1}:\n{text}\n\n"
    
    return extracted_text

# Example usage
pdf_path = "example.pdf"
extracted_text = extract_text_from_pdf(pdf_path)
print(extracted_text)
```

```python
#with word counting 
import PyPDF2
from transformers import pipeline
import re

def extract_text_from_pdf(pdf_path):
    # Initialize the AI-based OCR pipeline
    ocr = pipeline("image-to-text", model="microsoft/trocr-base-handwritten")

    # Open the PDF file
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        
        extracted_text = ""
        
        for page_num in range(num_pages):
            page = reader.pages[page_num]
            
            # Extract text using PyPDF2
            text = page.extract_text()
            
            # If PyPDF2 fails to extract text, use AI-based OCR
            if not text.strip():
                # Note: In a real implementation, you'd need to convert the PDF page to an image here
                # For demonstration, we'll assume this step and use a placeholder
                image = f"page_{page_num}.png"  # Placeholder
                text = ocr(image)[0]['generated_text']
            
            extracted_text += f"Page {page_num + 1}:\n{text}\n\n"
    
    return extracted_text

def count_words(text):
    # Remove special characters and digits
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    
    # Split the text into words
    words = text.split()
    
    # Count the words
    word_count = len(words)
    
    return word_count

# Example usage
pdf_path = "example.pdf"
extracted_text = extract_text_from_pdf(pdf_path)
word_count = count_words(extracted_text)

print(extracted_text)
print(f"\nTotal word count: {word_count}")
```
